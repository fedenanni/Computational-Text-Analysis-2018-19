```python
# so: how many tweets written by Donald Trump (and only by him!) contain the word "great"?

import json

with open('../datasets/trump.json') as f:
    tweets = json.load(f)

print (len(tweets))

written_by_trump = 0
with_great = 0
for tweet in tweets:
    if "Android" in tweet["source"]:
        text = tweet["text"].lower().split(" ")
        written_by_trump += 1
        if "great" in tweet["text"].lower() and "great" not in text:
            with_great += 1
            print (tweet["text"])
        
print (with_great, with_great/len(tweets),with_great/written_by_trump)
```
