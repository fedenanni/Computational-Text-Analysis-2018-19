```python
# opening text files
import codecs

speech = codecs.open("speech.txt","r","utf-8")

speech = speech.read()

# loop over files in folder
```

```python
words = speech.split(" ")
print (words)
```

```python
# list comprehension

words = [x.lower() for x in words]
```

```python
from collections import Counter

counting_words = Counter(words)

print counting_words.most_common(30)
```
